```python
d = {'text': [], 'label': []}

df = pd.DataFrame(data = d)
```

```python
sentences = []
labels = []
topics = ['environment', 'social programs', 'social cultural', 'foreign policy', 
          'immigration', 'economic', 'military','governance', 'safety'] 
import csv
with open("one_hot_fb_post.csv", 'r',encoding="utf8") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader)
    for row in reader:
        #print(row[1],row[0])
        for i in range(1,len(topics)+1):
            label = int(row[i])
            if label == 1:
                sentences.append(row[0])
                labels.append(topics[i-1])
        if int(row[-1]) == 0:
                sentences.append(row[0])
                labels.append('None')
 
t = {'text': sentences, 'label': labels}
t1 = pd.DataFrame(data = t)
df = df.append(t1, ignore_index = True)
        
```

```python
df.head()
```

```python
df.label = df.label.replace({'social programs':'sp'}, regex=True)
df.label = df.label.replace({'social cultural':'sc'}, regex=True)
df.label = df.label.replace({'foreign policy':'fp'}, regex=True)
df.to_csv('xx1.csv')
```
